Question: 

Draw a Decision Tree by the following data and information
Here,

Gain (Weather) = 0.124

Gain (Humidity) = 0.6

Gain (Wind) = 0

```python
from sklearn.tree import DecisionTreeClassifier, export_text, plot_tree
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
df
```

```python
df['Humidity'] = df['Humidity'].map({'High': 0, 'Normal': 1, 'Low': 2})
df['Play'] = df['Play'].map({'No': 0, 'Yes': 1})
```

```python
X = df[['Humidity']]
y = df['Play']
```

```python
clf = DecisionTreeClassifier(criterion="entropy", max_depth=1)
clf.fit(X, y)
```

```python
clf.predict([[0]])
```

```python
# Plot the decision tree
plt.figure(figsize=(10, 6))
plot_tree(clf, feature_names=['Humidity'], class_names=['No', 'Yes'], filled=True)
plt.show()
```

```python
plt.figure(figsize=(10, 7))
plot_tree(clf, 
          feature_names=['Weather', 'Humidity', 'Wind'], 
          class_names=['No', 'Yes'], 
          filled=True, 
          rounded=True)
plt.title("Decision Tree Visualization", fontsize=16)
plt.show()
```
