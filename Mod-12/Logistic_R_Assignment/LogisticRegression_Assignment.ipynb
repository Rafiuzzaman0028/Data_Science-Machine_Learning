**load the csv**

**Split the train and test data**

```python
#code here
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[['Pclass','Fare']], df.Survived, train_size=0.7)
```

```python
X_train
```

```python
X_test
```

**Make the model Binary logistic regression model**

```python
#code here
from sklearn.linear_model import LogisticRegression
blr = LogisticRegression()
blr.fit(X_train, y_train)
```

```python
y_test
```

```python
y_predicted = blr.predict(X_test)
y_predicted
```

```python
# For pclass 1 with Fare 55 dollars the passangers survived as model prediction
blr.predict([[1,55]])
```

```python
# For pclass 3 with Fare 8.0 dollars the passangers won't survived as model prediction
blr.predict([[3,8.0]])
```

**Check the model score**

```python
#code here
blr.score(X_test,y_test)
#the model accuracy is 66 % here
```
